```python
from googleapiclient.discovery import build
import json
import re
from tqdm.notebook import tqdm

API_KEY = 'AIzaSyBhT7i0HvxQ0JXWZHcBPMKrekE83dxvcxM'
SHEET1 = '1DKRq6CnpFm_wFnVVo2UnUfzuH_s2DtDsJMDAMnHYIFI'
SHEET2 = '1kboYe0wPftOOXp5NdnC5qP4iZzeUGVRMVKgul77AneQ'

def get_markup(table_row):
    print(len(table_row))
    data_object = {}
    if len(table_row) > 1:
        data_object['gunshot'] = table_row[1]
    else:
        return data_object
    if len(table_row) > 2:
        data_object['speech'] = table_row[2]
    else:
        return data_object
    if len(table_row) > 3:
            data_object['sound_from_target'] = table_row[3]
    else:
        return data_object
    if len(table_row) > 6:
            data_object['environment'] = table_row[6]
    else:
        return data_object
    if len(table_row) > 7:
            data_object['environment_tag'] = table_row[7]
    else:
        return data_object
    if len(table_row) > 8:
        data_object['distance_to_object'] = table_row[8]
    else:
        return data_object
    if len(table_row) > 9:
        data_object['target_material'] = table_row[9]
    else:
        return data_object
    if len(table_row) > 10:
        data_object['microphone_distance'] = table_row[10]
    else:
        return data_object
    if len(table_row) > 11:
        data_object['shooting_type'] = table_row[11]
    else:
        return data_object

with open('gunshot_data') as file:
    data = json.load(file)

service = build('sheets', 'v4', developerKey=API_KEY) 

result = service.spreadsheets().get(spreadsheetId=SHEET1).execute()
values = result.get('sheets', [])

ranges_template = ['!' + str(i) + ':' + str(i) for i in range(1, 1001)]
sheets = [value['properties']['title'] for value in values]

tables = []
for i, sheet in zip(range(6), sheets):
    range_names = [sheet + template for template in ranges_template]
    result = service.spreadsheets().values().batchGet(spreadsheetId=SHEET2, ranges=range_names).execute()
    for res in result['valueRanges']:
        if 'values' in res:
            tables += res['values']

regexp = re.compile(r'.+embed/(.+)\?start=(\d+)&end=(\d+)')
for row in tables:
    match = regexp.match(row[0])
    if match:
        if match.group(1) not in data:
            data[match.group(1)] = []
        data[match.group(1)].append(get_markup(row))

with open('gunshot_data', 'w') as file:
    data = json.dump(data, file, indent=4)
```

```python
def take_n_unique(dataset, n):
    viewed = {}
    result = []
    regexp = re.compile(r'.+embed/(.+)\?.*')
    for row in dataset:
        video_id = regexp.match(row[8]).group(1)
        if video_id not in viewed:
            viewed[video_id] = 1
            result.append(row)
        elif viewed[video_id] < n:
            viewed[video_id] += 1
            result.append(row)
    return result
```

```python
from googleapiclient.errors import HttpError

def check_availability(dataset, api_key):
    youtube = build('youtube', 'v3', developerKey=api_key)
    regexp = re.compile(r'.+embed/(.+)\?.*')
    for row in tqdm(dataset):
        match = regexp.match(row[8])
        if match:
            video_id = match.group(1)
            if len(row) < 10:
             try:
                 results = youtube.videos().list(id=video_id, part='status').execute()
                 if results['pageInfo']['totalResults']:
                     row.append('available')
                 else:
                     row.append('unavailable')
             except HttpError as err:
                 print(err)
                 break
        else:
            print(row)
```
